## Chinese Restaurant Process

```python
import numpy as np
from numpy.random import choice

def chinese_restaurant_process(n, alpha):
    if n < 1:
        return None
    
    table_assign = np.empty(n)
    next_table = 0
    
    for i in range(n):
        if np.random.random() < (1. * alpha / (alpha + i)):
            # sit at new table
            table_assign[i] = next_table
            next_table += 1
        else :
            # selection probs
            prob = [(table_assign[:i]==j).sum()/float(i)
                    for j in range(next_table)]
            # assign to existing table with prob
            table_assign[i] = choice(range(next_table),p=prob)
    return table_assign

print chinese_restaurant_process(100,1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
n = 100
alpha = 1

def plot_crp(n_table, ax=None):
    x = range(int(n_table.max()) + 1)
    y = [(n_table==i).sum() for i in set(n_table)]
    if ax is None: ax = plt
    ax.bar(x, y)

fig, axes = plt.subplots(10, 10, sharex=True, sharey=True, figsize=(10,6))
for ax in np.ravel(axes):
    plot_crp(chinese_restaurant_process(n, alpha), ax=ax)
```
